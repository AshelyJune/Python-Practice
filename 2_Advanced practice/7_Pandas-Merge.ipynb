```python
#importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country']== 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index= 'job_posted_month', columns= 'job_title_short',aggfunc='size')
df_US_pivot.reset_index(inplace= True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format = '%B').dt.month
df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)

df_US_pivot.drop(columns='month_no', inplace= True)

df_US_pivot
```

```python
# fake data to merge

df_US_pivot_software = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
```

```python
df_US_merged = df_US_pivot.merge(df_US_pivot_software, on= 'job_posted_month')
df_US_merged
```

```python
top_5 = df_US_merged.sum().sort_values(ascending=False).head(5).index.to_list()
```

```python
top_5
```

```python
df_US_merged[top_5].plot(kind='line')
plt.legend()
plt.ylim(0,20000)
plt.title("top_5 jobs")
plt.xlabel("2023")
```
